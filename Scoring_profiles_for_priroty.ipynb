```python
import pandas as pd
```

```python
swipes_df = pd.read_csv('/content/drive/MyDrive/swipes.csv', low_memory=False)
```

```python
id = 9
likes_performed = 0
likes_received = 0
dislikes_received = 0
successful_matches = 0
for i in range(len(swipes_df)): 
# print(swipes_df.loc[i, 'p1'])
# for j in range(10):
  if swipes_df.iloc[i]['p1'] == id:
    if(swipes_df.iloc[i]["first_type"] == 'LIKE'):
      likes_performed += 1
      if(swipes_df.iloc[i]["second_type"] == 'LIKE' and swipes_df.loc[i, "is_unmatch"] == False):
        successful_matches += 1


  elif(swipes_df.loc[i, 'p2'] == id):
    if(swipes_df.loc[i, "first_type"] == 'LIKE'):
      likes_received += 1
      if(swipes_df.iloc[i]["second_type"] == 'LIKE' and swipes_df.loc[i, "is_unmatch"] == False):
        successful_matches += 1
      
    else:
      dislikes_received += 1

```

```python

print('\nStats for user ', id)
print('Likes Performed: ', likes_performed)
print('likes_received: ', likes_received)
print('dislikes received: ', dislikes_received)
print('matches Received: ', successful_matches)
```
